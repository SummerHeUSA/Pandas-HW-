```python
df['Price_2'] = df['Price']
df.head()
```

```python
#purchasing analysis - gender# 
by_gender=df.groupby('Gender').agg({'Purchase ID':'count','Price':'mean','Price_2':'sum'}).reset_index()
by_gender["Price_per_purchase"]=by_gender["Price_2"]/by_gender["Purchase ID"]
by_gender
```

```python
#age demographics#

bins = [0, 10, 14, 18, 22, 26,30,34,38,42,46,50]
# labels = ['<10', '10-14','14-18','18-22','22-26','26-30','30-34','34-38','>38'  ]
df['binned'] = pd.cut(df['Age'], bins)
```

```python
df.head()
```

```python
by_age_group = df.groupby('binned').agg({'Purchase ID':'count','Price':'mean','Price_2':'sum'})

```

```python
by_age_group["Price_per_purchase_by_age"] = by_age_group["Price_2"]/by_age_group["Purchase ID"]
```

```python
by_age_group
```
